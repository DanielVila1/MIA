# 2. Model toyLM_LSTM

## 2.1. HerMajestySpeech Dataset

```python
# Define and train the tokenizer
tokenizer_m = Tokenizer(oov_token=0)
tokenizer_m.fit_on_texts(training_text.split("\n"))
```

```python
# Define the model
max_seq_length = 10
majesty_lstm = TextGenerator(tokenizer_m, max_seq_length)
```

```python
# TRAIN DATA
contexts_padded, one_hot_labels = majesty_lstm.generate_sequences(training_text.split("\n"))

# VALIDATION DATA
val_contexts_padded, val_one_hot_labels = majesty_lstm.generate_sequences(validation_text.split("\n"))

# TEST DATA
test_contexts_padded, test_one_hot_labels = majesty_lstm.generate_sequences(testing_text.split("\n"))
```

```python
# LSTM model
majesty_lstm.build(embedding_size=30, lstm_size=128, summary=True)
```

```python
# Train the model 
majesty_lstm.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
majesty_lstm.train_model(contexts=contexts_padded, labels=one_hot_labels, contexts_val=val_contexts_padded, labels_val=val_one_hot_labels, epochs=20, batch_size=128)
```

```python
# Evaluate the model
test_sentences = testing_text.split("\n")
perplexity = majesty_lstm.calculate_perplexity(test_sentences)
print("Test perplexity: ", perplexity)
```

```python
# Generate sentences
for _ in range(10):
    rdm_context = majesty_lstm.generate_context(testing_text.split("\n"))
    text = majesty_lstm.generate_text(rdm_context, 10, strategy='a')

    print("Context: ", rdm_context)
    print("Generated text: ", text)
    print("----------------------------------------------")
```

```python
tokenizer_b = Tokenizer(oov_token=0)
tokenizer_b.fit_on_texts(train_text + [' '])
```

```python
# LSTM model
book_lstm.build(embedding_size=30, lstm_size=128, summary=True)
```

```python
book_lstm.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
book_lstm.train_model(contexts=contexts_train, labels=labels_train, contexts_val=contexts_val, labels_val=labels_val, epochs=20, batch_size=128)
```

```python
# Evaluate the model
perplexity = majesty_lstm.calculate_perplexity(test_text)
print("Test perplexity: ", perplexity)
```
