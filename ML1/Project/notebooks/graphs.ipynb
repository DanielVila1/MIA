```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ann = [0.957, 0.945, 0.965, 0.930, 0.912, 0.840, 0.947, 0.927, 0.823, 0.915, 0.899, 0.866]
dt = [0.919, 0.908, 0.946, 0.867, 0.851, 0.775, 0.929, 0.910, 0.7,81, 0.821, 0.807, 0.786]
svm = [0.822, 0.820, 0.838, 0.739, 0.736, 0.596, 0.740, 0.737, 0.563, 0.747, 0.746, 0.742]
knn = [0.971, 0.958, 0.971, 0.956, 0.937, 0.860, 0.970, 0.950, 0.851, 0.945, 0.927, 0.877]

hardVoting = [0.968, 0.956, 0.969, 0.951, 0.933, 0.849, 0.960, 0.942, 0.832, 0.944, 0.928, 0.883]
softVoting = [0.969, 0.956, 0.968, 0.954, 0.934, 0.847, 0.959, 0.940, 0.824, 0.949, 0.931, 0.886]
randomForest = [0.864, 0.863, 0.942, 0.753, 0.749, 0.747, 0.931, 0.929, 0.864, 0.668, 0.664, 0.672]

test = [2,5,8,11]
ann_t = []
dt_t = []
svm_t = []
knn_t = []
hardVoting_t = []
softVoting_t = []
randomForest_t = []

for t in test:
    ann_t.append(ann[t])
    dt_t.append(dt[t])
    svm_t.append(svm[t])
    knn_t.append(knn[t])
    hardVoting_t.append(hardVoting[t])
    softVoting_t.append(softVoting[t])
    randomForest_t.append(randomForest[t])
```

```python
plotdata = pd.DataFrame({
    "ANN": ann_t,
    "SVM": svm_t,
    "DT": dt_t,
    "KNN": knn_t,
    },

    index=["Accuracy", "F-Score", "Precision", "Recall"])

color_dict = {'ANN': 'red', 'SVM': 'green', 'DT' : 'blue', 'KNN' : 'orange'}
plotdata.plot(color=[color_dict.get(x, '#333333') for x in plotdata.columns], kind="bar",figsize=(15, 8))


plt.ylabel('%', rotation = 0)
plt.grid(axis = 'y', color = 'gray', linestyle = 'dashed' )
plt.xticks(rotation=0)

plt.legend(loc = 'upper right')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
plotdata = pd.DataFrame({
    "Hard Voting": hardVoting_t,
    "Soft Voting": softVoting_t,
    "Random Forest": randomForest_t,
    "KNN": knn_t,
    },

    index=["Accuracy", "F-Score", "Precision", "Recall"])

color_dict = {'Hard Voting': 'purple', 'Soft Voting': 'cyan', 'Random Forest' : 'black', 'KNN' : 'orange'}
plotdata.plot(color=[color_dict.get(x, '#333333') for x in plotdata.columns], kind="bar",figsize=(15, 8))


plt.ylabel('%', rotation = 0)
plt.grid(axis = 'y', color = 'gray', linestyle = 'dashed' )
plt.xticks(rotation=0)

plt.legend(loc = 'upper right')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
